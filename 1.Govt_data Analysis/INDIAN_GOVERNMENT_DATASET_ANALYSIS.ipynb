# Dataset Description:
The dataset contains information about pensioners across 28 regions (states or territories). The columns present in the data are:

1. State Name: Name of the state or territory.
2. Total Pensioner: Total number of pensioners in the state.
3. Normal Pensioner: Pensioners under the normal pension scheme.
4. Family Pensioner: Pensioners receiving family pensions.
5. CDA Pensioner: Pensioners under the CDA (Central Dearness Allowance) scheme.
6. CDA Male Pensioner: Male pensioners under the CDA scheme.
7. CDA Female Pensioner: Female pensioners under the CDA scheme.
8. IDA Pensioner: Pensioners under the IDA (Industrial Dearness Allowance) scheme.
9. IDA Male Pensioner: Male pensioners under the IDA scheme.
10. IDA Female Pensioner: Female pensioners under the IDA scheme.

Each column contains numerical data (integers), except for the "State Name" column, which contains string values. There are no missing values in any of the columns.

# Dataset Observation:


### 1. Total Pensioners Across States
   - The total number of pensioners varies significantly across different states and territories. For example, Andhra Pradesh has the highest number of pensioners (20,007), while Andaman and Nicobar has the lowest (185).
  
### 2. Normal vs. Family Pensioners
   - In most states, the number of **Normal Pensioners** (pensioners under regular pension schemes) is higher than **Family Pensioners** (pensioners receiving family pensions).
   - For example, in Bihar, there are 8,510 normal pensioners compared to 3,917 family pensioners.

### 3. CDA vs. IDA Pensioners
   - Pensioners are categorized under two different schemes:
     - **CDA Pensioners**: Those under the Central Dearness Allowance scheme.
     - **IDA Pensioners**: Those under the Industrial Dearness Allowance scheme.
   - Generally, the number of IDA pensioners is higher than CDA pensioners in most states. For instance, Andhra Pradesh has 17,513 IDA pensioners compared to 2,494 CDA pensioners.
  
### 4. Gender Breakdown in CDA and IDA Schemes
   - Both CDA and IDA pensioners are further divided by gender.
   - Across the dataset, there is a clear disparity in the number of male and female pensioners in both schemes. Male pensioners significantly outnumber female pensioners in both the CDA and IDA categories.
     - For example, in Assam, there are 1,098 male CDA pensioners compared to just 41 female CDA pensioners. Similarly, there are 5,904 male IDA pensioners compared to 461 female IDA pensioners.

### 5. State-Wise Distribution
   - Larger states tend to have more pensioners. For example, states like Andhra Pradesh, Bihar, and Assam have a significantly higher number of pensioners across both CDA and IDA schemes.
   - Smaller states or territories, such as Andaman and Nicobar, have fewer pensioners overall.

### 6. CDA vs. IDA Proportions
   - In most cases, IDA pensioners dominate over CDA pensioners. This could imply a larger industrial workforce in those regions compared to government or central organizations that follow the CDA scheme.
   - For example, Bihar has 8,486 IDA pensioners versus 3,941 CDA pensioners, highlighting this trend.

### Summary:
- There is a significant variation in the number of pensioners across different states.
- Normal pensioners outnumber family pensioners across all regions.
- IDA pensioners generally outnumber CDA pensioners, indicating a more prominent industrial workforce in these regions.
- There is a noticeable gender disparity, with more male pensioners than female pensioners in both CDA and IDA schemes.
